```python
browser = webdriver.Chrome('./chromedriver.exe')
```

```python
browser.get('https://www.melon.com/chart/index.htm')
```

```python
html = browser.page_source
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
tags = soup.select('span > a')
tags
```

```python
result = []
results = []
for i in range(len(tags)-3) :
    if i % 2 == 0 :
        result = []
        title = tags[i].text.strip()
        result.append(title)
    else :
        name = tags[i].text.strip()
        link = tags[i]['title']
        result.append(name)
        result.append(link)
        results.append(result)
results
```
