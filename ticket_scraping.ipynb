```python
from selenium import webdriver
```

```python
browser = webdriver.Chrome('./chromedriver.exe')
```

```python
browser.get('http://ticket.interpark.com/Contents/Ranking?smid1=s_menu&smid2=performance')
```

```python
html = browser.page_source
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
tags = soup.select('div.rankingGenre')
```

```python
tags[0]
```

장르 : div.genreTitle
제목 : a.prdName
기간 : a.prdDuration

```python
tags[0].select('div.genreTitle')[0].text.strip()
```

```python
tags[0].select('a.prdName')[0].text.strip()
```

```python
tags[0].select('a.prdDuration')[0].text.strip()
```

```python
contents = []
for tag in tags :
    genre = tag.select('div.genreTitle')[0].text.strip()
    for i in range(len(tag)) :
        name = tag.select('a.prdName')[i].text.strip()
        dur = tag.select('a.prdDuration')[i].text.strip()
        content = [genre, name, dur]
        contents.append(content)
contents
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(contents, columns=['Genre','Title', 'Duration'])
df
```

```python
df.to_excel('./saves/ticket_scraping.xls',index=False)
```
